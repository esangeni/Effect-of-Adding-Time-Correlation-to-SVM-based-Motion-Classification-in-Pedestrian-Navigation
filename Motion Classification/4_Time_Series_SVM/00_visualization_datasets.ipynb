```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## **<font size="5">2. Parameters to modify</font>**
---

```python
data_dataset = '2022_03_17' # '2022_03_17' '2022_04_25' '2022_06_02' '2022_07_27' '2022_08_01'
type_dataset = '6sig'       # '2sig' '6sig' 'HD'
test_dataset = True         # False/True


dataset_path = 'G:\\Shared drives\\NIST-Pedestrian Activity Classification\\motion classification\\2_labeled_datasets\\'+data_dataset+'\\'
if data_dataset == '2022_06_02': dataset_path = 'G:\\Shared drives\\NIST-Pedestrian Activity Classification\\motion classification\\1_unlabeled_datasets\\2022_06_02_IEEE_Sensors_Conference_Paper\\'
```

---
## **<font size="5">3. Load Test Dataset:</font>**
---

```python
# Load 2sig Train/Test datasets
if type_dataset == '2sig':
    if test_dataset == True:
        exp_test_df = pd.read_csv(dataset_path + '2sig_test_df.csv')
    else:
        exp_test_df = pd.read_csv(dataset_path + '2sig_train_df.csv')
    print('1')

# Load 6sig Train/Test datasets
if type_dataset == '6sig':
    if test_dataset == True:
        exp_test_df = pd.read_csv(dataset_path + '6sig_test_df.csv')
    else:
        exp_test_df = pd.read_csv(dataset_path + '6sig_train_df.csv')
    print('2')

# Load HD Train/Test datasets
if type_dataset == 'HD':
    if test_dataset == True:
        exp_test_df = pd.read_csv(dataset_path + 'HD_200_2sig_test_df.csv')
    else:
        exp_test_df = pd.read_csv(dataset_path + 'HD_200_2sig_train_df.csv')
    print('3')

# Load Test Paper dataset:
if data_dataset == '2022_06_02' and type_dataset == '2sig':
    exp_test_df = pd.read_csv(dataset_path + '2022_06_02_2sig_test_df.csv')
    print('3')
if data_dataset == '2022_06_02' and type_dataset == 'HD':
    exp_test_df = pd.read_csv(dataset_path + '2022_06_02_HD_200_2sig_test_df.csv')
    print('4')



exp_test_df.head(2)
```

```python
if 'Unnamed: 0' in exp_test_df:
    exp_test_df = exp_test_df.drop(columns=['Unnamed: 0']) # Unnamed: 0

if 'Unnamed: 0.1' in exp_test_df:
    exp_test_df = exp_test_df.drop(columns=['Unnamed: 0.1']) # Unnamed: 0

if 'timestamp' in exp_test_df:
    exp_test_df = exp_test_df.drop(columns=['timestamp']) # Unnamed: 0

exp_test_df
```

---
## **<font size="5">4. Visualization:</font>**
---

<font size="2">Plot the 2 signals datasets:</font>

```python
if type_dataset == '2sig':
    plt.figure()
    sns.set(rc = {'figure.figsize':(15,8)})
    sns.lineplot(data = exp_test_df, x = exp_test_df.index, y = 'acc', hue=exp_test_df['label'])
    plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
    plt.xlim(0,max(exp_test_df.index))

    plt.figure()
    sns.set(rc = {'figure.figsize':(15,8)})
    sns.lineplot(data = exp_test_df, x = exp_test_df.index, y = 'gyro', hue=exp_test_df['label'])
    plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
    plt.xlim(0,max(exp_test_df.index))
```

<font size="2">Plot the 6 signals datasets:</font>

```python
if type_dataset == '6sig':
    plt.figure()
    sns.set(rc = {'figure.figsize':(15,8)})
    sns.lineplot(data = exp_test_df, x = exp_test_df.index, y = 'acc3', hue=exp_test_df['label'])
    plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
    plt.xlim(0,max(exp_test_df.index))

    plt.figure()
    sns.set(rc = {'figure.figsize':(15,8)})
    sns.lineplot(data = exp_test_df, x = exp_test_df.index, y = 'gyro3', hue=exp_test_df['label'])
    plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
    plt.xlim(0,max(exp_test_df.index))
```

<font size="2">Plot the HD 2 signals datasets:</font>

```python
if type_dataset == 'HD':
    plt.figure()
    sns.set(rc = {'figure.figsize':(15,8)})
    sns.lineplot(data = exp_test_df, x = exp_test_df.index, y = 'acc_t0')
    plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
    plt.xlim(0,max(exp_test_df.index))

    plt.figure()
    sns.set(rc = {'figure.figsize':(15,8)})
    sns.lineplot(data = exp_test_df, x = exp_test_df.index, y = 'gyro_t0')
    plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
    plt.xlim(0,max(exp_test_df.index))
```
